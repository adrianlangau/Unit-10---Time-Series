```python
# Forcasting Returns using an ARMA Model
# Create a series using "Price" percentage returns, drop any nan"s, and check the results:
# (Make sure to multiply the pct_change() results by 100)
# In this case, you may have to replace inf, -inf values with np.nan"s
returns = (cad_jpy_df[["Price"]].pct_change() * 100)
returns = returns.replace(-np.inf, np.nan).dropna()
returns.tail()
```

```python
import statsmodels.api as sm
from statsmodels.tsa.arma_model import ARMA
model = ARMA(returns.values, order=(2,1))
```

```python
# Forcasting Returns using an ARIMA Model
from statsmodels.tsa.arima.model import ARIMA
```

```python
model = ARIMA(returns.values, order=(5,1,1))
```

```python
results = model.fit()
```

```python
# Print the model summary
results.summary()
```

```python
# Plot the Futures Price Forecast
pd.DataFrame(results.forecast(steps=5)[0]).plot(title="5 Day Futures Price Forecast")
```

```python
# Volatility Forecasting with GARCH
from arch import arch_model
```

```python
# create the GARCH Model
model = arch_model(returns, mean="Zero", vol="GARCH", p=2, q=1)
```

```python
results = model.fit(disp="off")
```

```python
results.summary()
```

```python
# Find the last day of the dataset
last_day = returns.index.max().strftime('%Y-%m-%d')
last_day
```

```python
# Create a 5 day forecast of volatility
forecast_horizon = 5

forecasts = results.forecast(start='2020-06-04', horizon=forecast_horizon)
```

```python
# Annualize the forecast
intermediate = np.sqrt(forecasts.variance.dropna() * 252)
intermediate.head()
```

```python
# Transpose the forecast so that it is easier to plot
final = intermediate.dropna().T
final.head()
```

```python
final.plot()
```
